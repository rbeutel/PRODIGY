```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
quad = pd.read_csv('ALBH_PROD_NA.neu', sep=" ", header=None)
albh = pd.read_csv('QUAD_PROD_NA.neu', sep=" ", header=None)
```

```python
header = ['time (decimal year)', 'north (mm)', 'east (mm)', 'up (mm)']
quad.columns = header
albh.columns = header
```

```python
# estimate the approximate azimuth of the particle velocity vector due to the long-term (~550 yr) locking/loading signal with respect to north and its rate in this direction in mm/yr.
x = quad['east (mm)']
y = quad['north (mm)']
m, b = np.polyfit(x, y, 1)
plt.scatter(x,y,s=1)
plt.plot(x, m*x + b,'k')
plt.ylabel('North (mm)')
plt.xlabel('East (mm)')
```

```python
angle  = 90-math.degrees(math.atan(m))
print("the the approximate azimuth of the particle velocity vector at QUAD is {}".format(angle))
```

```python
nproj = quad['north (mm)']/math.cos(m)
eproj = quad['east (mm)']/math.sin(m)
total = eproj + nproj
```

```python
y = total
x = quad['time (decimal year)']
m, b = np.polyfit(x, y, 1)
plt.scatter(x,y, s=1)
plt.plot(x, m*x + b,'k')
```

```python
print("The slope at QUAD is equal to {}, mm/yr".format(m))
```
