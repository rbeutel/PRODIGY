```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
lonlat = pd.read_csv('SSTlonlat.txt', sep=" ")
land = pd.read_csv('SSTlandmask.txt', sep=" ")
data = pd.read_csv('SSTdata.txt', sep=" ")
```

```python
data['landmask'] = land['mask']
data['lon'] = lonlat['lon']
data['lat'] = lonlat['lat']
data
```

## Preprocessing
Delete land rows and make seperate datasets of up to April 1997 and just October 1997

```python
#delete land rows
data = data[data['landmask'] != 1]
```

```python
data
```

```python
sst = data[data.columns[0:328]]
octoberSST = data[data.columns[334]]
```

## Compute EOF (principal components)

```python
sst = sst.to_numpy()
```

```python
Z = sst.T
```

```python
spat_mean = np.mean(Z, 0)
np.shape(spat_mean)
```

```python
np.shape(Z)
```

```python
Zspat_detrend = [Z[i,:] - spat_mean for i in range(np.shape(Z)[0])]
Zspat_detrend = np.asarray(Zspat_detrend)
# Zspat_detrend = Zspat_detrend.T
```

```python
np.shape(Zspat_detrend)
```

```python
nt = 328 #number of observations
Zt = Zspat_detrend / math.sqrt(nt - 1)
```

```python
U, S, Ev = np.linalg.svd(Zt)
Ev = Ev.T
```

#### pick 10 EOFs for this analysis

```python
n = 10;
TS = np.matmul(Zspat_detrend, Ev[:, 1:n]) #can't just straight up multiply matrices in python, must use this function
```

```python
# lets check a day to see how good out reconstruction is!
SSTreconstructed = np.matmul(Ev[:, 1:n],TS[228, :].T) + spat_mean.T
# got to plot this to compare but not the focus rn
```

```python
plt.plot(Z[228,:], label='original')
plt.plot(SSTreconstructed, label='reconstructed')
plt.legend()
```

## Dynamic Model

```python
tau = 6;
# [nt d] = size(TS); nt still 328?
Cov0 = np.matmul(TS.T,TS) / nt 
TStplustau = TS[(tau+1):nt, :] #TS with first tau time pts removed
TSt = TS[1:(nt-tau),:]    # TS with last tau time pts removed
Covtau = np.matmul(TSt.T, TStplustau).T / (nt - tau)
```

```python
# In MATLAB?
C0inv = np.linalg.inv(Cov0) #<- solve(Cov0)
Mest = Covtau * C0inv
Ceta = Cov0 - Covtau * C0inv * Covtau.T
```

### Compute prediction 6 months ahead

```python
alpha_forecast = Mest * TS[327, :].T
```

```python
spatial_forecast = [np.matmul(Ev[:, 1:n],alpha_forecast)[:,i] + spat_mean.T for i in range(np.shape(alpha_forecast)[0])]
```

```python
C = Mest * Cov0 * Mest.T + Ceta;
```

```python
forecast_bounds = 2 * np.sqrt(np.diag(np.matmul(np.matmul(Ev[:, 1:n], C), Ev[:, 1:n].T)))
```

```python
plt.plot(octoberSST, label='original')
plt.plot(forecast_bounds, label='forecasted')
plt.legend()
```
